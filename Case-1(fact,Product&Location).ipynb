```python
# #22. Write a function by creating a condition in which if the total expenses is less than
# 60 then it is a profit or else loss.
fact['Result'] = np.where(fact['Total Expenses'] < 60, 'Profit', 'Loss')
fact.head()

def profit_or_loss(tot_expense):
    if tot_expense < 60: 
        return "Profit"
    else:
        return "Loss"

fact['Result'] = fact['Total Expenses'].apply(profit_or_loss)
fact.head()
```

```python
# 23. Give the total weekly sales value with the date and product ID details. 

Table['week'] = pd.to_datetime(Table['Date']).dt.isocalendar().week
Table.columns
weekly_sales_per_productID = Table.groupby(['week','ProductId'])['Sales'].sum()
weekly_sales_per_productID.head()

####some advanced levels 
levels = ['week', 'Date', 'ProductId']
results = []

for i in range(len(levels), -1, -1):
    current_levels = levels[:i]
    if current_levels:
        grouped = Table.groupby(current_levels)['Sales'].sum().reset_index()
    else:
        # Grand Total layer
        grouped = pd.DataFrame([{'Sales': Table['Sales'].sum()}])
        
    results.append(grouped)

comb_table = pd.concat(results, ignore_index=True).sort_values(by=levels, na_position='last')
comb_table.head(20)
```
